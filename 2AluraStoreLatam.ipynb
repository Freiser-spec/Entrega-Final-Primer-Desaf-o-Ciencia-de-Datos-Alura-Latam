### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular el ingreso total de la Tienda 1
ingreso_tienda1 = tienda['Precio'].sum()
print(f"Ingreso total de Tienda 1: ${ingreso_tienda1:,.2f}")

# Calcular el ingreso total de la Tienda 2
ingreso_tienda2 = tienda2['Precio'].sum()
print(f"Ingreso total de Tienda 2: ${ingreso_tienda2:,.2f}")

# Calcular el ingreso total de la Tienda 3
ingreso_tienda3 = tienda3['Precio'].sum()
print(f"Ingreso total de Tienda 3: ${ingreso_tienda3:,.2f}")

# Calcular el ingreso total de la Tienda 4
ingreso_tienda4 = tienda4['Precio'].sum()
print(f"Ingreso total de Tienda 4: ${ingreso_tienda4:,.2f}")

# Resumen en un diccionario:
ingresos_por_tienda = {
    "Tienda 1": ingreso_tienda1,
    "Tienda 2": ingreso_tienda2,
    "Tienda 3": ingreso_tienda3,
    "Tienda 4": ingreso_tienda4
}

print("\nIngresos totales por tienda (diccionario):")
print(ingresos_por_tienda)
```

```python
import matplotlib.pyplot as plt

# Datos de facturación total por tienda (usando el diccionario que creamos antes)
tiendas = list(ingresos_por_tienda.keys())
facturacion = list(ingresos_por_tienda.values())

# 3. Gráfico de barras con anotaciones de los valores
plt.figure(figsize=(10, 7))
bars = plt.bar(tiendas, facturacion, color=['aquamarine', 'goldenrod', 'aqua', 'crimson'])
plt.xlabel("Tienda")
plt.ylabel("Facturación Total (USD)")
plt.title("Facturación Total por Tienda con Valores Anotados")
plt.grid(axis='y', linestyle='--')

# Anotar los valores en las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1000, '${:,.2f}'.format(yval), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
def identificar_productos_extremos(df, nombre_tienda):
    """
    Identifica los productos más y menos vendidos en una tienda y muestra los resultados como texto.

    Args:
        df (pd.DataFrame): DataFrame de la tienda.
        nombre_tienda (str): Nombre de la tienda.
    """
    if 'Producto' not in df.columns:
        print(f"Advertencia: La columna 'Producto' no se encontró en {nombre_tienda}.")
        return

    conteo_productos = df['Producto'].value_counts()
    if conteo_productos.empty:
        print(f"Advertencia: No hay datos de productos en {nombre_tienda}.")
        return

    mas_vendido = conteo_productos.index[0]
    cantidad_mas_vendido = conteo_productos.iloc[0]
    menos_vendido = conteo_productos.index[-1]
    cantidad_menos_vendido = conteo_productos.iloc[-1]

    print(f"\n--- Resultados de {nombre_tienda} ---")
    print(f"Producto Más Vendido: {mas_vendido} (Cantidad: {cantidad_mas_vendido})")
    print(f"Producto Menos Vendido: {menos_vendido} (Cantidad: {cantidad_menos_vendido})")

# Aplicar la función a cada tienda
identificar_productos_extremos(tienda, "Tienda 1")
identificar_productos_extremos(tienda2, "Tienda 2")
identificar_productos_extremos(tienda3, "Tienda 3")
identificar_productos_extremos(tienda4, "Tienda 4")
```

```python
import matplotlib.pyplot as plt

def visualizar_ventas_con_marcadores(df, nombre_tienda):
    """
    Visualiza los productos más y menos vendidos con marcadores.

    Args:
        df (pd.DataFrame): DataFrame de la tienda.
        nombre_tienda (str): Nombre de la tienda para el título del gráfico.
    """
    if 'Producto' not in df.columns:
        print(f"Advertencia: La columna 'Producto' no se encontró en {nombre_tienda}.")
        return

    conteo_productos = df['Producto'].value_counts()
    if conteo_productos.empty:
        print(f"Advertencia: No hay datos de productos en {nombre_tienda}.")
        return

    mas_vendido = conteo_productos.index[0]
    cantidad_mas_vendido = conteo_productos.iloc[0]
    menos_vendido = conteo_productos.index[-1]
    cantidad_menos_vendido = conteo_productos.iloc[-1]

    plt.figure(figsize=(8, 6))
    productos = ['Más Vendido', 'Menos Vendido']
    cantidades = [cantidad_mas_vendido, cantidad_menos_vendido]
    nombres = [mas_vendido, menos_vendido]
    colores = ['dodgerblue', 'firebrick']
    marcadores = ['o', '^'] # Círculo para más vendido, triángulo para menos vendido

    for i, producto in enumerate(productos):
        plt.plot(producto, cantidades[i], marker=marcadores[i], markersize=15, color=colores[i], label=f'{producto}: {nombres[i]}')
        plt.vlines(producto, 0, cantidades[i], linestyles='dashed', color=colores[i], alpha=0.7)
        plt.text(producto, cantidades[i] + 1, f'({cantidades[i]})', ha='center', va='bottom')

    plt.ylabel("Cantidad Vendida")
    plt.title(f"Productos Más y Menos Vendidos en {nombre_tienda}")
    plt.xticks(productos)
    plt.legend()
    plt.grid(axis='y', linestyle='--')
    plt.tight_layout()
    plt.show()

# Aplicar la función a cada tienda
visualizar_ventas_con_marcadores(tienda, "Tienda 1")
visualizar_ventas_con_marcadores(tienda2, "Tienda 2")
visualizar_ventas_con_marcadores(tienda3, "Tienda 3")
visualizar_ventas_con_marcadores(tienda4, "Tienda 4")
```

```python
import matplotlib.pyplot as plt

def visualizar_ventas_geograficas(df, nombre_tienda):
    """
    Visualiza las ventas geográficas de una tienda usando un gráfico de dispersión.

    Args:
        df (pd.DataFrame): DataFrame de la tienda con columnas 'lat' y 'lon'.
        nombre_tienda (str): Nombre de la tienda para el título del gráfico.
    """
    if 'lat' not in df.columns or 'lon' not in df.columns:
        print(f"Advertencia: Las columnas 'lat' o 'lon' no se encontraron en {nombre_tienda}.")
        return

    plt.figure(figsize=(10, 8))
    plt.scatter(df['lon'], df['lat'], alpha=0.5, label=nombre_tienda)
    plt.xlabel("Longitud")
    plt.ylabel("Latitud")
    plt.title(f"Distribución Geográfica de Ventas en {nombre_tienda}")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Aplicar la función a cada tienda
visualizar_ventas_geograficas(tienda, "Tienda 1")
visualizar_ventas_geograficas(tienda2, "Tienda 2")
visualizar_ventas_geograficas(tienda3, "Tienda 3")
visualizar_ventas_geograficas(tienda4, "Tienda 4")
```

```python
import folium

def visualizar_ventas_folium(df, nombre_tienda):
    """
    Visualiza las ventas geográficas de una tienda en un mapa interactivo de Folium.

    Args:
        df (pd.DataFrame): DataFrame de la tienda con columnas 'lat' y 'lon'.
        nombre_tienda (str): Nombre de la tienda para el título del mapa.
    """
    if 'lat' not in df.columns or 'lon' not in df.columns:
        print(f"Advertencia: Las columnas 'lat' o 'lon' no se encontraron en {nombre_tienda}.")
        return

    # Calcular la ubicación promedio para centrar el mapa (opcional)
    lat_promedio = df['lat'].mean()
    lon_promedio = df['lon'].mean()

    # Crear el mapa
    m = folium.Map(location=[lat_promedio, lon_promedio], zoom_start=10)

    # Añadir marcadores para cada venta
    for index, row in df.iterrows():
        folium.Marker([row['lat'], row['lon']], popup=f"Venta en {row['lat']:.4f}, {row['lon']:.4f}").add_to(m)

    # Guardar el mapa como un archivo HTML
    map_filename = f'mapa_ventas_{nombre_tienda}.html'
    m.save(map_filename)
    print(f"Mapa interactivo de {nombre_tienda} guardado como {map_filename}")

# Aplicar la función a cada tienda
visualizar_ventas_folium(tienda, "Tienda 1")
visualizar_ventas_folium(tienda2, "Tienda 2")
visualizar_ventas_folium(tienda3, "Tienda 3")
visualizar_ventas_folium(tienda4, "Tienda 4")
```
